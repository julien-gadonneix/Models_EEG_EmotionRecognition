```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/Users/juliengadonneix/Desktop/stage 3a/data/LUMED-2/EEG_GSR_Data/'
mat = scipy.io.loadmat(path + 's01.mat')
print(mat.keys())
print(mat['__header__'])
print(mat['__version__'])
print(mat['__globals__'])
```

```python
eeg = mat['EEGData'][0, 0]
print(len(eeg))
print(eeg[0].shape)
print(eeg[1].shape)
```

```python
sr = 500
tf = eeg[0].shape[0]/sr
print(tf//60, 'min', tf%60, 's')
time = np.linspace(0, tf, eeg[0].shape[0])

fig, axes = plt.subplots(3, 3, figsize=(15, 10))
for i in range(3):
    for j in range(3):
        axes[i,j].plot(time, eeg[0][:, j+3*i])
        axes[i,j].set_xlabel('Time (s)')
        axes[i,j].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(time, eeg[1])
plt.xlabel('Time (s)')
plt.xticks(np.arange(0, tf, 20))
plt.ylabel('Classes')

plt.show()
```
